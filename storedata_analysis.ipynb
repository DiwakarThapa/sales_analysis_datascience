# # FINDING OVERALL SALES TREND

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['year_month'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['year_month']
```

```python
#grouping the year month for sales
df_salestrends = df.groupby('year_month').sum()['sales'].reset_index()
df_salestrends
```

```python
# plotting the chart.
plt.figure(figsize = (18,8))
plt.plot(df_salestrends['year_month'],df_salestrends['sales'])
plt.xticks(rotation = 'vertical')
plt.show()
```

# Top product by sales

```python
product_list_sales = pd.DataFrame(df.groupby('product_name').sum()['sales'])
product_list_sales
```

```python
# sorting the dataframe
sorted_product_list = product_list_sales.sort_values(by = ['sales'],ascending = False)
```

```python
sorted_product_list[:10]
```

# Finding most selling product

```python
prod_quantity = pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
prod_quantity_df = prod_quantity.sort_values(by = ['quantity'],ascending = False)[:10]
```

```python
prod_quantity_df
```

# Finding most preferred shipping mode

```python
ship_mode_df = df['ship_mode'].value_counts().reset_index()
ship_mode_df
```

```python
plt.figure(figsize= (18,8))
sns.countplot(df['ship_mode'])
plt.show()
```

# Most profitable category and sub-category

```python
category_df = pd.DataFrame(df.groupby(['category','sub_category']).sum()['profit'])
```

```python
category_df.sort_values(['category','profit'],ascending = False)
```

```python
category_df.plot(kind = 'bar',figsize = (18,8))
plt.show()
```
